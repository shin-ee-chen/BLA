# 1. Active/Passive Random Examples

Dataset size: 613

Demo example number: 5

```python
n_demo = 5
```

```python
active_passive_captions = utils.read_json_file('data/active_passive.json')
len(active_passive_captions)
```

```python
sent_len = []
voc_dict = defaultdict(lambda:int(0))

for img in active_passive_captions:
    captions = img['caption_group'][0]
    sents = [captions['True1'], captions['True2'], \
        captions['False1'], captions['False2']]
    for sent in sents:
        sent_len.append(len(sent.split()))
        for word in sent.split():
            voc_dict[lemma(word)] += 1
print("average sentence length:", np.mean(sent_len))
print("vocaulary size:", len(voc_dict))
print(sorted(voc_dict.items(), key = lambda x: x[1], reverse=True)[:10])
```

```python
selects = np.random.randint(0, len(active_passive_captions), n_demo)
```

```python
for idx in selects:
    item = active_passive_captions[idx]
    img_id = item['image_id']
    img = item['caption_group'][0]
    utils.show_image(img_id)
    print(img_id)
    print(img['True1'])
    print(img['True2'])
    print(img['False1'])
    print(img['False2'])
```
